# Handling Missing Data
Missing data exists for several reasons, could be because of inconsistencies in data collection/recording, entry errors or simply not enough data etc... 

This step, there are other packages you can use like sklearn.impute.SimpleImputer etc, but I am more conversant with handling missing data with pandas!

Ultimately when handling missing data, there are 2 main possible outcomes:
1. Drop columns/rows with missing data
2. Impute values

But there are advantages and disadvantages for each of these methods (NOT LIMITED TO, THESE ARE BASED ON MY EXPERIENCE):

#### Dropping columns/rows with missing data:

+: Easy, fast, saves (A LOT) of time.

-: Dataset becomes more susceptible to biasness (more pronounced in smaller datasets, can be negligible for (VERY) large datasets (>50K observations), you may end up deleting other valuable feature observations along with the missing data within the row. 

#### Imputation:

+: Preserving size of dataset and observations, valuable feature observations is retained.

-: Time consuming (But can be very worth it if your dataset is small and you want to preserve as much observations as you possibly can).

3. inplace= True, will permenantly reflect the change on the DataFrame. Default value is inplace= False.

```python
# method= 'ffill' - Forward Fill
df3.fillna(method= 'ffill')
```

Missing data will be filled with data from the same column in the preceding row, as simple as data!

```python
# method= 'bfill' - Backward Fill
df3.fillna(method= 'bfill')
```
